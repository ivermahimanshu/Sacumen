```python
# Program 1

max=1
p=0
q=0
list_of_max=[]
def iterate(a,temp_list):
    if a in temp_list:
        global max,q,p
        q=a
        max+=1
        return iterate(a+1,temp_list)
    else:
        print(p,",",q)
        
def return_range_output(temp2_list):
    i=temp2_list[0]
    if i in temp2_list:
        global p,q
        p=i
        if i+1 in temp2_list:
            iterate(i+1,temp2_list)
        else:
            list_of_max.append
            return_range_output(temp2_list[temp2_list.index(i)+1:])        
list_of_numbers=list(map(int,input().split()))
list_of_numbers.sort()
return_range_output(list_of_numbers)
```

```python
#Program 3

from collections import Counter
import string
def count_words(name):
    with open(name,"r") as f:
        count=0
        lst=f.readlines()
        for i in lst:
            words=i.split(" ")
            count+=len(words)
        print("Number of words in the file is : ",count)
def count_dupe_words(name):
    with open(name,"r") as f:
        count=0
        s=f.read()
        lst=s.split()
        #print(string.punctuation)
        for i in lst:
            if i in string.punctuation:
                lst.remove(i)
        d=Counter(lst)
        for k,v in d.items():
            if v>1:
                count+=1
        print("Number of duplicate words are : ",count)
def count_special_chars(name):
    with open(name,"r") as f:
        count=0
        s=f.read()
        lst=s.split()
        for i in lst:
            if i in string.punctuation:
                count+=1
            for j in i:
                if j in string.punctuation:
                    count+=1
        print("Number of special characters in the file Is : ",count)
def count_spaces(name):
    with open(name,"r") as f:
        count=0
        lst=f.readlines()
        for i in lst:
            words=i.split())
            count=count+len(words)-1
        print("Number of spaces in the file is : ",count)
    
file_name=input("Enter the path of the file : ")
count_words(file_name)
count_dupe_words(file_name)
count_special_chars(file_name)
count_spaces(file_name)
```

```python
#Program 4

import random
import array
min_len = 8
digits = ['2', '3', '4', '5', '6', '7', '8', '9']  
locase_characters = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 
                    'j', 'k', 'm', 'n', 'p', 'q',
                     'r', 's', 't', 'u', 'v', 'w', 'x', 'y',
                     'z']
  
upcase_characters = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 
                     'I', 'J', 'K', 'M', 'N', 'P', 'Q',
                     'R', 'S', 'T', 'U', 'V', 'W', 'X', 'Y',
                     'Z']
  
symbols = ['@', '#', '$', '%', '=', '?', '|', '~', 
           '*']

combines_list = digits + upcase_characters + locase_characters + symbols
rand_digit = random.choice(digits)
rand_upper = random.choice(upcase_characters)
rand_lower = random.choice(locase_characters)
rand_symbol = random.choice(symbols)

temp_pass = rand_digit + rand_upper + rand_lower + rand_symbol
for x in range(min_len - 4):
    temp_pass = temp_pass + random.choice(combines_list)
    temp_pass_list = array.array('u', temp_pass)
    random.shuffle(temp_pass_list)

password = ""
for x in temp_pass_list:
        password = password + x
print(password)
```
